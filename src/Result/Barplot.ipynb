```python

 import pandas as pd
```

```python
df=pd.read_csv('final_mean_metrics.csv')
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df
```

```python
mod=list(df['models'])
```

```python
combine_acc=list(df['combine_acc'])
```

```python
news_acc=list(df['news_acc'])
```

```python
numerical_acc=list(df['numerical_acc'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

mod[0]="Log Reg"
 
data = {mod[i]: combine_acc[i] for i in range(len(mod))}

models = list(data.keys())
accuracies = list(data.values())

fig = plt.figure(figsize = (8, 5))

plt.bar(models, accuracies, color ='royalblue',width = 0.2)
		
plt.ylim([min(combine_acc)-0.01, max(combine_acc)+0.01])

plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Model vs Accuracy for combined data")
plt.show()
```
